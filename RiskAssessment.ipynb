```
plt.suptitle("Agglomerative Clustering",fontsize=20)


model = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
X = df.loc[:,["Result", "Country_Region"]].values
model.fit_predict(X)
labels = model.labels_

plt.figure(figsize=(24,4))
plt.subplot(1,3,1)
plt.title("Clusters = 2",fontsize=16)
plt.xlabel("Confirmed Cases/Population")
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50)
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50)
plt.show()



model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
X = df.loc[:,["Result", "Country_Region"]].values
model.fit_predict(X)
labels = model.labels_

plt.figure(figsize=(24,4))
plt.subplot(1,3,2)
plt.title("Clusters=5",fontsize=16)
plt.xlabel("Confirmed Cases/Population")
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, color = 'green')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, color = 'blue')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=50, color = 'purple')
plt.scatter(X[labels==3, 0], X[labels==3, 1], s=50, color = 'orange')
plt.scatter(X[labels==4, 0], X[labels==4, 1], s=50, color = 'red')
plt.show()



model = AgglomerativeClustering(n_clusters=6, affinity='euclidean', linkage='ward')
X = df.loc[:,["Result", "Country_Region"]].values
model.fit_predict(X)
labels = model.labels_

plt.figure(figsize=(24,4))
plt.subplot(1,3,3)
plt.title("Clusters = 6",fontsize=16)
plt.xlabel("Confirmed Cases/Population")
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, color = 'green')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, color = 'blue')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=50, color = 'orange')
plt.scatter(X[labels==3, 0], X[labels==3, 1], s=50, color = 'purple')
plt.scatter(X[labels==4, 0], X[labels==4, 1], s=50, color = '')
plt.scatter(X[labels==5, 0], X[labels==5, 1], s=50)
plt.show()
```

```
#Trying a different way with regularized data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
results_scaled = scaler.fit_transform(df[['Result', 'Country_Region']])
# statistics of scaled data
df2 =  pd.DataFrame(results_scaled)
df2.columns = ['Result',  'Country_Region']
df2.describe()
```

```
#inertia  on the  fitted data
kmeans.inertia_
```

```
# fitting multiple k-means algorithms and storing the values in an empty list
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster, init='k-means++')
    kmeans.fit(results_scaled)
    SSE.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```
